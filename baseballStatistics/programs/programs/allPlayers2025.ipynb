Saving all players from the 2025 Season

```python
# Takes the file's folder
filepath = r'c:\Users\jvgas\OneDrive\Desktop\learning\github\baseballStatistics\py4da\baseballStatistics\datasets\retrosheet data\allPlayers\allplayers.csv';

# read the CSV file
allPlayers = pd.read_csv(filepath)

# trimming all allPlayers to only 2025 season
allPlayers2025 = allPlayers.loc[allPlayers['season'] == 2025]

# renaming columns of allPlayers2025
allPlayers2025.rename(columns = {'id' :'RetroID' , 'last' : 'Last name', 'first' : 'First name', 'bat' : 'Bat Side',
                                                  'throw' : 'Throw Side', 'team' : 'Team ID', 'g' : 'Games Played', 'g_p' : 'Games as Pitcher',
                                                  'g_sp' : 'Games at Starting Pitcher', 'g_rp' : 'Games at Relief Pitcher',
                                                  'g_c' : 'Games at Catcher', 'g_1b' : 'Games at 1B', 'g_2b' : 'Games at 2B', 'g_3b' : 'Games at 3B',
                                                  'g_ss' : 'Games at SS', 'g_lf' : 'Games at LF', 'g_cf' : 'Games at CF', 'g_rf' : 'Games at RF',
                                                  'g_of' : 'Games at OF', 'g_dh' : 'Games at DH', 'g_ph' : 'Games at Pinch Hitter',
                                                  'g_pr' : 'Games at Pinch Runner', 'first_g' : 'First Game Played',
                                                  'last_g' : 'Last Game Played', 'season' : 'Season'}, inplace = True)
allPlayers2025['Full Name'] = allPlayers2025['First name'] + ' ' + allPlayers2025['Last name']

# saving allPlayers2025 to csv

# Change CWD  
os.chdir(save_dir)  
 
# Verify the new CWD  
allPlayers2025.to_csv('Players2025.csv')
print('File saved successfully in the specified directory:', save_dir)
```
