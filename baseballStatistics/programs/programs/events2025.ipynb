```python
# Importiang statcast data function call
from pybaseball import statcast 

# 2025 Statcast Data Retrieval
events2025 = statcast(start_dt="2025-01-01", end_dt="2025-11-11")


# Renaming colums for events2025
events2025.rename(columns = {'pitch_type' :'Pitch Type' , 'game_date' : 'Game Date', 'release_speed' : 'Release Speed', 'release_pos_x' : 'X Release',
                             'release_pos_z' : 'Z Release', 'player_name' : 'Player Name', 'batter' : 'Batter ID', 'pitcher' : 'Pitcher ID',
                             'events' : 'Events', 'description' : 'Pitch Description', 'spin_dir' : 'Spin Direction', 'spin_rate_deprecated' : 'Spin Rate (Deprecated)',
                             'break_angle_deprecated' : 'Break Angle (Deprecated)', 'break_length_deprecated' : 'Break Length (Deprecated)',
                             'zone' : 'Zone', 'des' : 'Play Description', 'game_type' : 'Game Type', 'stand' : 'Stand', 'p_throws' : 'Pitcher Hand',
                             'home_team' : 'Home Team', 'away_team' : 'Away Team', 'type' : 'Pitch Result', 'hit_location' : 'Hit Location', 'bb_type' : 'Batted Ball Type',
                             'balls' : 'Pre-pitch # of Balls', 'strikes' : 'Pre-pitch # of Strikes', 'game_year' : 'Year', 'pfx_x' : 'X Movement (feet)', 
                             'pfx_z' : 'Z Movement (feet)', 'plate_x' : 'X Position at Plate', 'plate_z' : 'Z Position at Plate', 'on_3b' : 'Runner ID on 3B',
                             'on_2b' : 'Runner ID on 2B', 'on_1b' : 'Runner ID on 1B', 'outs_when_up' : '# of Outs', 'inning' : 'Inning', 'inning_topbot' : 'Inning TopBot',
                             'hc_x' : 'X Position of Batted Ball', 'hc_y' : 'Y Position of Batted Ball', 'tfs_deprecated' : 'Deprecated field from old tracking system',
                             'tfs_zulu_deprecated' : 'Deprecated field from old tracking system', 'fielder_2' : 'Player ID of Catcher', 'umpire' : 'Deprecated field from old tracking system',
                             'sv_id' : 'Play Event ID', 'vx0' : 'Pitch X Velocity', 'vy0' :'Pitch Y Velocity' , 'vz0' : 'Pitch Z Velocity', 'ax' : 'Pitch X Acceleration',
                             'ay' : 'Pitch Y Acceleration', 'az' : 'Pitch Z Acceleration', 'sz_top' : 'Strike Zon Top', 'sz_bot' : 'Strike Zone Bottom', 'hit_distance' : 'Hit Distance',
                             'launch_speed' : 'Exit Velocity', 'launch_angle' : 'Launch Angle', 'effective_speed' : 'Effective Spin', 'release_spin' : 'Spin Rate', 
                             'release_extension' : 'Release Extension', 'game_pk' : 'Game ID', 'fielder_2' : 'Catcher ID', 'fielder_3' : '1B ID', 'fielder_4' : '2B ID', 'fielder_5' : '3B ID', 
                             'fielder_6' : 'SS ID', 'fielder_7' : 'LF ID', 'fielder_8' : 'CF ID', 'fielder_9' : 'RF ID', 'release_pos_y' : 'Y Release', 
                             'estimated_ba_using_speedangle' : 'Estimated Batting Average', 'estimated_woba_using_speedangle' : 'Estimated wOBA', 'wOBA_value' : 'wOBA Value',
                             'wOBA_denom' : 'wOBA Denominator', 'babip_value' : 'BABIP Value', 'iso_value' : 'ISO Value', 'launch_speed_angle' : 'Launch Speed Angle', 
                             'at_bat_number' : 'At Bat Number', 'pitch_number' : 'Pitch Number', 'pitch_name' : 'Pitch Name', 'home_score' : 'Home Team Score', 'away_score' : 'Away Team Score', 
                             'bat_score' : 'Batting Team Score', 'fld_score' : 'Fielding Team Score', 'post_home_score' : 'Post-Pitch Home Team Score', 'post_away_score' : 'Post-Pitch Away Team Score',
                             'if_fielding_alignment' : 'IF Alignment', 'of_fielding_alignment' : 'OF Alignment', 'spin_axis' : 'Spin Axis', 'delta_home_win_exp' : 'Change in Home Team Win Expectancy',
                             'delta_run_exp' :'Change in Run Expectancy', 'bat_speed' :'Bat Speed' , 'swing_length' : 'Swing Length', 'estimated_slg_using_speed_angle' : 'Estimated SLG.', 
                             'delta_pitcher_run_exp' : 'Change in Pitcher Run Expectancy', 'hyper_speed' : 'Hyper Speed', 'home_score_diff' : 'Home Score Differential',
                             'batt_score_diff' : 'Batting Team Score Differential', 'home_win_exp' : 'Home Team Win Expectancy', 'age_pit_legacy' : 'Pitcher Age (Legacy)',
                             'age_bat_legacy' : 'Batter Age (Legacy)', 'age_pit' : 'Pitcher Age', 'age_bat' : 'Batter Age', 'n_thruorder_pitcher' : 'Pitcher Times Thru Order',
                             'n_priorpa_thisgame_player_at_bat' : 'Batter PA This Game', 'pitcher_days_since_prev_game' : 'Pitcher Days Since Last Appearance', 
                             'batter_days_since_prev_game' : 'Batter Days Since Last Appearance', 'pitcher_days_until_next_game' : 'Pitcher Days Until Next Appearance',
                             'batter_days_until_next_game' : 'Batter Days Until Next Appearance', 'api_break_z_with_gravity' : 'Drop', 'api_break_x_arm' : 'X Break (Arm)', 'api_break_x_batter_in' : 'X Break (Batter)',
                             'arm_angle' : 'Arm Angle', 'attack_direction' : 'Attack Direction', 'effective_spin_efficiency' : 'Effective Spin Efficiency', 'swing_path_tilt' : 'Swing Path Tilt', 
                             'intercept_ball_minus_batter_pos_x_inches' : 'X Position at Intercept', 'intercept_ball_minus_batter_pos_y_inches' : 'Y Position at Intercept', 'post_bat_score' : 'Post-Pitch Batting Team Score'}, inplace = True)

# Calculating Spray Angle and Defining Hit Field

events2025['Spray Angle'] = np.round((np.arctan((events2025['X Position of Batted Ball'] - 130) / (213 - events2025['Y Position of Batted Ball'])) * 180 / pi * 0.75), 1)
events2025['Stand'] = events2025['Stand'].astype(str)

events2025['Spray Angle'] = events2025['Spray Angle'].fillna(9999)

def classifyHitFieldRight(events2025):
    if -50 <= events2025 < -10:
        hitField = 'PULL'
        return hitField
    elif -10 <= events2025 < 10:
        hitField = 'STRAIGHTAWAY'
        return hitField
    elif 10 <= events2025 < 50:
        hitField = 'OPPO'
        return hitField

def classifyHitFieldLeft(events2025):
    if -50 <= events2025 < -10:
        hitField = 'OPPO'
        return hitField
    elif -10 <= events2025 < 10:
        hitField = 'STRAIGHTAWAY'
        return hitField
    elif 10 <= events2025 < 50:
        hitField = 'PULL'
        return hitField


if events2025['Stand'].eq('R').all():
    events2025['Hit Field'] = events2025['Spray Angle'].apply(classifyHitFieldRight)
else:
    events2025['Hit Field'] = events2025['Spray Angle'].apply(classifyHitFieldLeft)



# saving events2025 to csv

# Change CWD  
os.chdir(data_dir)  
 
# Verify the new CWD  
events2025.to_csv('events2025.csv')
print('File saved successfully in the specified directory:', data_dir)
```
