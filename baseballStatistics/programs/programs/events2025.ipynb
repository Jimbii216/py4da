```python
# Calculate Spray Angle
events2025['Spray Angle'] = np.round((np.arctan((events2025['X Position of Batted Ball'] - 130) / (213 - events2025['Y Position of Batted Ball'])) * 180 / pi * 0.75), 1)

# Create Hit Direction column based on batter stance
# For RIGHT-handed batters: Left Pull (-50 to -15), Center Straight (-15 to 15), Right Oppo (15 to 50)
# For LEFT-handed batters: Right Pull (15 to 50), Center Straight (-15 to 15), Left Oppo (-50 to -15)

conditions = [
    (events2025['Stand'] == 'R') & (events2025['Spray Angle'] >= -50) & (events2025['Spray Angle'] < -15),
    (events2025['Stand'] == 'R') & (events2025['Spray Angle'] >= -15) & (events2025['Spray Angle'] < 15),
    (events2025['Stand'] == 'R') & (events2025['Spray Angle'] >= 15) & (events2025['Spray Angle'] < 50),
    (events2025['Stand'] == 'L') & (events2025['Spray Angle'] >= -50) & (events2025['Spray Angle'] < -15),
    (events2025['Stand'] == 'L') & (events2025['Spray Angle'] >= -15) & (events2025['Spray Angle'] < 15),
    (events2025['Stand'] == 'L') & (events2025['Spray Angle'] >= 15) & (events2025['Spray Angle'] < 50),
]

choices = ['Left', 'Center', 'Right', 'Right', 'Center', 'Left']

events2025['Hit Direction'] = np.select(conditions, choices, default='Unknown')

print(events2025[['Spray Angle', 'Stand', 'Hit Direction']].head(20))
```
