```python
import pandas as pd
import numpy as np
import os  
 
# Get the current working directory  
current_dir = os.getcwd()  
 
print("Current Working Directory:", current_dir)

save_dir = r'C:\Users\jvgas\OneDrive\Desktop\learning\github\baseballStatistics\py4da\baseballStatistics\datasets\jimbo datasets\trimmed retrosheet data\2025'
print('')
print("Current Saving Directory:", save_dir)
```

```python
# Takes the file's folder
filepath = r'C:\Users\jvgas\OneDrive\Desktop\learning\github\baseballStatistics\py4da\baseballStatistics\datasets\retrosheet data\batting\batting.csv';

# read the CSV file
allBatting = pd.read_csv(filepath)

# trimming Batting to only 2025 season
batting2025 = allBatting.loc[allBatting['date'] >= 20250101]


# renaming columns
batting2025.rename(columns = {'gid' : 'Game ID', 'id' : 'Player ID', 'team' : 'Team ID', 'b_lp' : 'Batter Lineup Position',
                                                  'b_seq' : 'Starter', 'stattype' : 'Stat Type', 'b_pa' : 'Plate Appearances', 'b_ab' : 'At Bats', 
                                                  'b_r': 'Runs Scored', 'b_h' : 'Hits', 'b_d' : 'Doubles', 'b_t' : 'Triples', 'b_hr' : 'Home Runs',
                                                  'b_rbi' : 'RBI', 'b_sh' : 'Sacrifice Hits', 'b_sf' : 'Sacrifice Flies' , 'b_hbp' : 'Hit by Pitch',
                                                  'b_w' : 'Walks', 'b_iw' : 'Intentional Walks' , 'b_k' : 'Strikeouts', 'b_sb' : 'Stolen Bases',
                                                  'b_cs' : 'Caught Stealing', 'b_gdp' : 'Ground into Double Play', 'b_xi' : ' Reached Bbase on Catchers Interference', 
                                                  'b_roe' : 'Reached Base on Error', 'dh' : 'DH', 'ph' : 'Pinch Hit', 'pr' : 'Pinch Run', 'date' : 'Date', 
                                                  'number' : 'Game Number (0 if Single Game)', 'site' : 'Site', 'vishome' : 'Home Team', 
                                                  'opp' : 'teams opponent in game', 'win' : 'Win', 'loss' : 'Loss', 'tie' : 'Tie', 
                                                  'gametype' : 'Game Type', 'box' : 'Box Score', 'pbp' : 'Play-By-Play'}, inplace = True)

# saving batting to csv

# Change CWD  
os.chdir(save_dir)  
 
# Saving and verifying correct save location   
batting2025.to_csv('batting2025.csv')
print('File saved successfully in the specified directory:', save_dir)
```
