##### Loading in the battingData file

```python
allBattingData = pd.read_excel('MLB Batting Data.xlsx')
```

```python
allBattingData['AVG'] = allBattingData['AVG'].replace({0:np.nan})
```

```python
hist = allBattingData['AVG'].hist(bins=100, color='skyblue', edgecolor='black')

plt.title('Histogram for Length Column') 
plt.xlabel('Length') 
plt.ylabel('Frequency') 

plt.show() 
```

```python
sns.histplot(allBattingData['AVG'], bins = 100, kde=True, color='lightgreen', edgecolor='black')

plt.xlabel('Values')
plt.ylabel('Density')
plt.title('2024 MLB Batting AVG Histogram with Density Plot')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))
 
axes[0].hist(allBattingData['AVG'], bins=100, color='Yellow', edgecolor='black')
axes[0].set_title('2024 MLB Batting Average')
 
axes[1].hist(allBattingData['HR'], bins=50, color='Pink', edgecolor='black')
axes[1].set_title('2024 MLB Home Runs')
 
# Adding labels and title
for ax in axes:
    ax.set_xlabel('Values')
    ax.set_ylabel('Frequency')
 
# Adjusting layout for better spacing
plt.tight_layout()
 
# Display the figure
plt.show()

Out
```
