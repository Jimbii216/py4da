```python
fantasyData = pd.read_excel('MLB Batting Data.xlsx', sheet_name = 'fantasyStats')
```

```python
fantasyData['TB'] = ((fantasyData['HR'] * 4) + (fantasyData['3B'] * 3) + (fantasyData['2B'] * 2) +  (fantasyData['1B'] * 1))
fantasyData['Walks'] = ((fantasyData['BB'] * 1) + (fantasyData['IBB']* 1))
```

```python
fantasyData['Points'] = ((fantasyData['R'] * 1) + (fantasyData['HR'] * 4) + (fantasyData['TB'] * 1) + (fantasyData['RBI'] * 1) + (fantasyData['Walks'] * 1) + (fantasyData['SO'] * (-1)) + (fantasyData['SB'] * 2))
```

```python
fantasyData.sort_values(by='Points', ascending=False)
```

```python
index_names = fantasyData[(fantasyData['PA'] <= 250)].index
fantasyDataTrimmed = fantasyData.drop(index_names)
```

```python
avg1 = fantasyDataTrimmed['Points'].mean(axis = 0)
avg1 = round(avg1)
avg1 = str(avg1)
print('The average is number of points scored in the 2024 Penthouse League was ' + avg1 + '!')
avg1 = int(avg1)
print()


avg2 = fantasyDataTrimmed['HR'].mean(axis = 0)
avg2 = round(avg2)
avg2 = str(avg2)
print('The average number of homerunes hit by players in 2024 was ' + avg2 + '!')
avg2 = int(avg2)
print()
print('##############')
print()

median1 = fantasyDataTrimmed['Points'].median(axis = 0)
median1 = round(median1)
median1 = str(median1)
print('The median of the number of points scored in the 2024 Penthouse League was ' + median1 + '!')
median1 = int(median1)
print()

median2 = fantasyDataTrimmed['HR'].median(axis = 0)
median2 = round(median2)
median2 = str(median2)
print('The median of the number of homeruns hit by players in 2024 was ' + median2 + '!')
median2 = int(median2)
print()
print('##############')
print()

series = pd.Series(fantasyDataTrimmed['Points'])

sd1 = series.std()
sd1 = round(sd1, 1)
strSD1 = str(sd1) 

print('The standard deviation in points scored is ' + strSD1 + '!')
print()
series = pd.Series(fantasyDataTrimmed['HR'])

sd2 = series.std()
sd2 = round(sd2)
strSD2 = str(sd2) 

print('The standard deviation in Home Runs hit is ' + strSD2 + '!')

### Three SD Lines ###

P1SD = np.round(median1 + sd1)
print(P1SD)
P2SD = np.round(P1SD + sd1)
print(P2SD)
P3SD = np.round(P2SD + sd1)
print(P3SD)

M1SD = np.round(median1 - sd1)
print(M1SD)
M2SD = np.round(M1SD - sd1)
print(M2SD)



P1SDHR = np.round(median2 + sd2)
print(P1SDHR)
P2SDHR = np.round(P1SDHR + sd2)
print(P2SDHR)
P3SDHR = np.round(P2SDHR + sd2)
print(P3SDHR)

M1SDHR = np.round(median2 - sd2)
print(M1SDHR)
M2SDHR = np.round(M1SDHR - sd2)
print(M2SDHR)
```

```python
fig, axes = plt.subplots()
 
axes.hist(fantasyDataTrimmed['Points'], color='Thistle', edgecolor='black')
axes.set_title('2024 Fantasy League Points Totals (Estimated)')
plt.axvline(x=median1, color='navy', label = 'Median') # Add Median Line
plt.axvline(x = P1SD, color = 'black', linestyle = '--', label = 'Standard Deviations')
plt.axvline(x = P2SD, color = 'black', linestyle = '--')
plt.axvline(x = P3SD, color = 'black', linestyle = '--')
plt.axvline(x = M1SD, color = 'black', linestyle = '--')
plt.axvline(x = M2SD, color = 'black', linestyle = '--')

plt.xlabel("Points")
plt.ylabel('Number of Players')
plt.legend()

# axes[1].hist(fantasyDataTrimmed['HR'], bins=50, color='Pink', edgecolor='black')
# axes[1].set_title('2024 MLB Home Runs')
# axes[1].plt.axvline(x=median2)

# # Adding labels and title
#for ax in axes:
 #    ax.set_xlabel('Values')
  ##   ax.set_ylabel('Frequency')
 
# # Adjusting layout for better spacing
# plt.tight_layout
# plt.plot(avg1)
# plt.plot(avg2) 
# # Display the figure
# plt.show()
```
